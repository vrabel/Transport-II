```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# define parameters
c10 = 1 # M
l = 1 # cm

def lowDa(z):
    Da = 0.01
    return c10*(1-z/l)

def Da(z,Da):
    Da = 5
    return c10*((1-(np.exp(np.sqrt(Da))/(2*np.sinh(np.sqrt(Da)))))*np.exp(np.sqrt(Da))*z/l+
                (np.exp(np.sqrt(Da)*(1-z/l)))/(2*np.sinh(np.sqrt(Da))))

def highDa(z):
    Da = 100
    delta = 0.01*l
    return c10*np.exp(-np.sqrt(Da)*z)
```

```python
z = np.linspace(0,l,1000)

plt.plot(z,lowDa(z),'--')
plt.plot(z,Da(z,5),'-.')
plt.plot(z,highDa(z))
plt.title('Concentration Profile for a Reactive Flim\n at Varying $Da_{II}$')
plt.legend(['$Da_{II}$ = 0.01','$Da_{II}$ = 5','$Da_{II}$ = 100'])
plt.xlabel('Distance (cm)')
plt.ylabel('Concentration (M)')
plt.axis([0,1,0,1])
```
